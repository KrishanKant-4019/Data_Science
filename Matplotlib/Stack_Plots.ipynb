# Stack Plot Example

```python
import matplotlib.pyplot as plt  # Import plotting library

days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]  # Hours spent studying each day
playing = [2, 2, 1, 1, 2, 3, 2]  # Hours spent playing each day
watching_tv = [2, 1, 2, 2, 1, 1, 1]  # Hours spent watching TV each day
sleeping = [5, 5, 6, 5, 6, 5, 5]  # Hours spent sleeping each day

labels = ["Studying", "Playing", "Watching TV", "Sleeping"]  # Labels for activities
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']  # Colors for stackplot areas

plt.figure(figsize = (6,6))  # Create a figure with size 6x6 inches
plt.stackplot(days, studying, playing, watching_tv, sleeping, labels = labels, colors = colors, alpha = 0.8)  # Draw stackplot for activities over days
plt.xlabel("Days")  # Label x-axis
plt.ylabel("Hours")  # Label y-axis
plt.title("Student's Activities Comparison")  # Title of the plot
plt.legend(loc = "upper left", framealpha = 0.5)  # Add legend with transparency
plt.grid(True, linestyle = "--", alpha = 0.7, color = "gray")  # Add grid with dashed lines and transparency
plt.show()  # Display the plot
```
