# Basic Scatter Plot

```python
import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]

plt.scatter(study_hours, exam_scores, color = "red")
plt.title("Study Hours vs Exam Scores", fontsize = 12, fontweight = "bold")
plt.xlabel("Study Hours", fontsize = 12)
plt.ylabel("Exam Scores", fontsize = 12)
plt.grid(True, linestyle = "--", alpha = 0.7, color = "black")
plt.show()
```

# Adding Color and Size

```python
size = [score * 2 for score in exam_scores]
colors = ["red" if score < 50 else "green" for score in exam_scores]

plt.scatter(study_hours, exam_scores, c = colors, s = size)
plt.xlabel("Study Hours", fontsize = 12)
plt.ylabel("Exam Scores", fontsize = 12)
plt.title("Study Hours vs Exam Scores", fontsize = 12, fontweight = "bold")
plt.grid(True, linestyle = "--", alpha = 0.7, color = "black")
plt.show()
```

 # Using a Colormap

```python
import numpy as np

scores_normalized = np.array(exam_scores)

plt.scatter(study_hours, exam_scores, c = scores_normalized, cmap = "viridis")
plt.colorbar(label = "Score")
plt.xlabel("Study Hours", fontsize = 12)
plt.ylabel("Exam Scores", fontsize = 12)
plt.title("Study Hours vs Exam Scores", fontsize = 12, fontweight = "bold")
plt.grid(True, linestyle = "--", alpha = 0.7, color = "black")
plt.show()
```

# Adding Annotations

```python
scores_normalized = np.array(exam_scores)

plt.scatter(study_hours, exam_scores, c = scores_normalized, cmap = "viridis")

for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}', (study_hours[i], exam_scores[i]))
 
plt.colorbar(label = "Score")
plt.xlabel("Study Hours", fontsize = 12)
plt.ylabel("Exam Scores", fontsize = 12)
plt.title("Study Hours vs Exam Scores", fontsize = 12, fontweight = "bold")
plt.grid(True, linestyle = "--", alpha = 0.7, color = "black")
plt.show()
```

# Multiple Groups in One Plot

```python
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='Class A', color='blue')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='orange')

plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
