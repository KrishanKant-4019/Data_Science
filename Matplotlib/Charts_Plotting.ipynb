# Real Data Example: Sachin Tendulkar’s Runs Over Time

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]  # List of years when runs were recorded
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]     # Corresponding runs scored in those years

plt.plot(years, runs)                     # Plot the data as a line graph (years on x-axis, runs on y-axis)
plt.xlabel("Year")                        # Label for the x-axis
plt.ylabel("Runs Scored")                # Label for the y-axis
plt.title("Sachin Tendulkar's Yearly Run")  # Title of the plot
plt.show()                                # Display the plot 
```

# Multiple Lines in One Plot

```python
Kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]  # Runs scored by Kohli in corresponding years
Rohit = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]  # Runs scored by Rohit in corresponding years

plt.plot(years, Kohli, label = "Virat Kohli")    # Plot Kohli's data with label
plt.plot(years, Rohit, label = "Rohit Sharma")   # Plot Rohit's data with label

plt.xlabel("Years")              # Label for the x-axis
plt.ylabel("Run Scored")         # Label for the y-axis
plt.title("Performance Comparison")  # Title of the plot
plt.legend()                     # Show the legend to distinguish players
plt.show()                       # Display the plot
```

# Using Format Strings

```python
plt.plot(years, Kohli, "ro--", label="Virat Kohli")   # Plot Kohli’s runs with red circles and dashed line, add label
plt.plot(years, Rohit, "g^:", label="Rohit Sharma")  # Plot Rohit’s runs with green triangles and dotted line, add label
plt.legend()                                         # Show legend to identify each player’s line
plt.show()                                           # Display the plot
```

# Color and Line Style Arguments(More Readable)

```python
plt.plot(years, Kohli, color="orange", linestyle="--", label="Virat Kohli")  # Plot Kohli with orange dashed line and label
plt.plot(years, Rohit, color="green", linestyle="-.", label="Rohit Sharma")  # Plot Rohit with green dash-dot line and label
plt.legend()  # Show legend to identify each player’s line
plt.show()    # Display the plot
```

# Line Width and Layout Tweaks

```python
plt.plot(years, Kohli, color="orange", linestyle="--", linewidth=3, label="Virat Kohli")  # Plot Kohli with orange dashed line, thicker width, and label
plt.plot(years, Rohit, color="green", linestyle="-.", linewidth=2, label="Rohit Sharma")  # Plot Rohit with green dash-dot line, medium width, and label
plt.legend()                  # Show legend to distinguish the two players
plt.grid(True)                # Display grid lines on the plot for better readability
plt.tight_layout()            # Adjust layout to prevent overlap of plot elements
plt.show()                   # Display the plot
```

# Visualizing Tons of Data 

```python
import numpy as np  # Import NumPy for generating random numbers

for i in range(50):  # Loop 50 times
    plt.plot(np.random.rand(100), linewidth=1)  # Plot 100 random values as a line with width 1

plt.title("Too Much Data Can Be Confusing!")  # Set the plot title
plt.grid(True)  # Show grid lines for better readability
plt.tight_layout()  # Adjust layout to prevent overlap of plot elements
plt.show()  # Display the final plot with all 50 lines
```
