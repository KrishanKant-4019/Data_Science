```python
import matplotlib.pyplot as plt
```

```python
# Setting a Style
plt.style.use("Solarize_Light2")
```

# Basic Pie Chart Example

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]  # player names
runs = [18000, 8000, 12000, 9500]                # corresponding run totals
colors = ["#7ecf50", "#5093cf", "#7150cf", "#bf1e1e"]  # custom slice colors

plt.title("Career runs of Indian Batsman")       # set chart title
plt.pie(runs, labels=labels, colors=colors)      # create pie chart with labels and colors
plt.show()                                       # display the chart
```

# Add Edges and Style Slices with wedgeprops

```python
plt.pie(runs,                                    # data for slices
        labels = labels,                         # labels for each wedge
        colors = colors,                         # custom colors
        wedgeprops = {                           # style properties for each wedge
            "edgecolor" : "black",               # border color
            "linewidth" : 1,                     # border width
            "linestyle" : "--",                  # dashed border
            # "width" : 0.8,                     # donut-style width (0 < width <= 1)
            "alpha" : 0.4,                       # transparency level
            # "hatch": "/"                       # diagonal hatch pattern
        })

plt.show()                                       # display the plot
```

# Highlight a Slice Using explode

```python
explode = [0.1, 0, 0, 0]                         # offset first slice by 0.1, others not exploded

plt.pie(runs,                                    # data values for pie slices
        labels = labels,                         # text labels for each slice
        colors = colors,                         # apply custom colors to slices
        shadow = True,                           # create shadow in pie chart
        explode = explode,                       # pull out first slice slightly (exploded effect)
        wedgeprops = {                           # styling properties for wedges
            "edgecolor": "black",                # color of slice borders
            "linewidth": 1,                      # thickness of slice borders
            # "linestyle": "--",                 # dashed style for slice borders
            # "width": 0.8,                      # donut hole width 
            "alpha": 0.4,                        # transparency level of slices
            # "hatch": "/"                       # hatch pattern
        })

plt.title("Exploded Pie Chart")                  # title of the plot
plt.show()                                       # render and display the pie chart
```

# Start Angle and Percentage

```python
plt.title("Career runs of Indian Batsman")      # set chart title
plt.pie(
    runs,                                       # data values for pie slices
    labels=labels,                              # text labels for each slice
    colors=colors,                              # apply custom colors to slices
    explode=explode,                            # offset slices as per explode list
    shadow=True,                                # add shadow for 3D effect
    startangle=180,                             # rotate start angle to 180 degrees
    autopct="%1.1f%%"                           # display percentage value on each slice with 1 decimal place
)
plt.show()                                      # render and display the pie chart
```
