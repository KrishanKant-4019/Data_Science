# Sachin’s Yearly Runs – Vertical Bar Chart

```python
# Basic Bar Plot
import matplotlib.pyplot as plt  # Import plotting module

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]  # X-axis data
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]     # Y-axis data

plt.bar(years, runs)  # Create bar chart
plt.xlabel("Year")  # X-axis label
plt.ylabel("Runs Scored")  # Y-axis label
plt.title("Sachin's Yearly Runs")  # Chart title
plt.tight_layout()  # Adjust layout
plt.show()  # Show the chart
```

# Setting Bar Width & Side-by-Side Bar Charts

```python
# Plotting Side-by-Side Bars
import numpy as np  # Import NumPy for numerical operations

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]  # Sachin's runs
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]        # Sehwag's runs
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]       # Kohli's runs

x = np.arange(len(years))  # Index positions for bars
width = 0.25  # Width of each bar(Default is 0.8)

plt.bar(x - width, sachin, width=width, label="sachin")  # Bars for Sachin (left)
plt.bar(x, sehwag, width=width, label="sehwag")          # Bars for Sehwag (center)
plt.bar(x + width, kohli, width=width, label="kohli")    # Bars for Kohli (right)

plt.xlabel("Year")  # X-axis label
plt.ylabel("Runs Scored")  # Y-axis label
plt.title("Performance Comparison")  # Chart title

plt.xticks(x, years)  # Set custom x-tick labels
plt.legend()  # Show legend

plt.tight_layout()  # Adjust layout
plt.show()  # Display chart
```

# Horizontal Bar Charts with barh()

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]  # Player names (Y-axis)
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]  # Total runs in 5 years

plt.barh(players, runs_5yrs, color="skyblue")  # Horizontal bar chart(y, width)

plt.xlabel("Total runs in 5 years")  # X-axis label
plt.title("First 5-Year Performance of Indian Batsmen")  # Chart title

plt.tight_layout()  # Adjust layout
plt.show()  # Show the plot
```

# Adding Value Labels with plt.text()

```python
import matplotlib.pyplot as plt  # Import plotting library

players = ["Sachin", "Sehwag", "Kohli"]  # Player names for X-axis
runs = [1500, 1200, 1800]  # Runs scored by each player

plt.bar(players, runs, color="skyblue")  # Create vertical bar chart with skyblue bars

for i in range(len(players)):  # Loop through each player index
    plt.text(i, runs[i] + 30, str(runs[i]), ha='center')  # Add runs value text above each bar, centered

plt.ylabel("Runs")  # Label Y-axis as "Runs"
plt.title("Runs Scored by Players")  # Add chart title

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plot
```
