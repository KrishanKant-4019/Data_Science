```python
# Question-1: Create a plot comparing Kohli, Rohit Sharma, and Sehwag across 10 years of hypothetical runs.
# Use:
# Labels
# Legends
# Colors
# Line styles
# One custom style
```

```python
import matplotlib.pyplot as plt  # Import matplotlib for plotting

plt.style.use("ggplot")  # Use ggplot style for the plot

years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]  # Years on X-axis
rohit_runs = [550, 700, 850, 900, 1100, 1400, 1000, 850, 750, 600]  # Rohit's runs per year
sehwag_runs = [450, 400, 200, 0, 0, 0, 0, 0, 0, 0]  # Sehwag's runs per year
virat_runs = [0, 20, 450, 650, 900, 1150, 1400, 1200, 1100, 850]  # Virat's runs per year

plt.plot(years, rohit_runs, color="blue", linestyle="--", linewidth=3, label="Rohit Sharma")  # Plot Rohit's runs with dashed blue line
plt.plot(years, virat_runs, color="red", linestyle="-.", linewidth=3, label="Virat Kohli")  # Plot Virat's runs with dash-dot red line
plt.plot(years, sehwag_runs, color="orange", linestyle=":", linewidth=3, label="Virendra Sehwag")  # Plot Sehwag's runs with dotted orange line

plt.xlabel("Year")  # Set X-axis label
plt.ylabel("Runs Scored")  # Set Y-axis label
plt.title("Year-Wise Performance Comparison")  # Set chart title
plt.grid(True)  # Enable grid for better readability
plt.legend()  # Show legend to identify lines
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plot
```

```python
# Question-2: Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.
```

```python
import matplotlib.pyplot as plt
import numpy as np  # Import numpy for array operations

years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009]  # List of years
sachin_runs = [1200, 1500, 1800, 1700, 1600, 1400, 1300, 1250, 1100, 1000]  # Sachin's runs per year
sehwag_runs = [800, 950, 1200, 1400, 1300, 1500, 1700, 1600, 1400, 1300]  # Sehwag's runs per year
kohli_runs = [1000, 800, 650, 900, 1400, 750, 800, 600, 500, 900]  # Kohli's runs per year

x = np.arange(len(years))  # Array of indices for the years
width = 0.25  # Width of each bar

plt.bar(x - width, sachin_runs, width=width, color="skyblue", label="Sachin Tendulkar")  # Sachin's bars shifted left
plt.bar(x, sehwag_runs, width=width, color="orange", label="Virendra Sehwag")  # Sehwag's bars centered
plt.bar(x + width, kohli_runs, width=width, color="red", label="Virat Kohli")  # Kohli's bars shifted right

plt.xlabel("Year")  # Label for X-axis
plt.ylabel("Runs Scored")  # Label for Y-axis
plt.title("Year-Wise Performance Comparison")  # Chart title
plt.xticks(x, years)  # Set X-axis ticks to years
plt.legend()  # Show legend identifying players
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the plot
```

```python
# Question-4: Plot a line graph to visualize the sales trend. Identify the month with the highest and lowest sales.
```

```python
import matplotlib.pyplot as plt

Month = ["January","February", "March", "April", "May", "June"]
Sales = [120, 150, 130, 170, 160, 180]

max_sales = max(Sales)
min_sales = min(Sales)
max_month = Month[Sales.index(max_sales)]
min_month = Month[Sales.index(min_sales)]
                  
plt.style.use("ggplot")

plt.plot(Month, Sales, label = "Sales", linestyle = "--", color = "red", linewidth = 2, marker = "o")

for i in range(len(Month)):
    plt.text(Month[i], Sales[i]+5, str(Sales[i]), fontsize = 9, ha = "center")
             
plt.annotate(f"Highest:{max_sales}", xy = (max_month, max_sales), xytext = (max_month, max_sales + 10),
             arrowprops = dict(facecolor = "green", shrink = 0.05), fontsize = 10, color = "green")
plt.annotate(f"Lowest:{min_sales}", xy = (min_month, min_sales), xytext = (min_month, min_sales - 10),
             arrowprops = dict(facecolor = "blue", shrink = 0.05), fontsize = 10, color = "blue")

plt.axhline(y=160, color = "gray", linestyle = "--", label = "Target Sales")

plt.xticks(rotation = 45)

plt.xlabel("Month")
plt.ylabel("Sales(units)")
plt.title("Monthly Sales")
plt.grid(True, linestyle = "--", alpha = 0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
 # Question-5: Plot a line chart to represent temperature changes throughout the morning.
```

```python
import matplotlib.pyplot as plt

# Time of day and corresponding temperatures
Time = ["6AM", "7AM", "8AM", "9AM", "10AM", "11AM", "12PM"]
Temperature = [15, 16, 18, 20, 22, 23, 25]

# Get highest and lowest temperatures and their corresponding times
max_temp = max(Temperature)
min_temp = min(Temperature)
max_time = Time[Temperature.index(max_temp)]
min_time = Time[Temperature.index(min_temp)]

# Set plot style
plt.style.use("Solarize_Light2")

# Plot the temperature line with markers
plt.plot(Time, Temperature, label="Temperature", color="green",
         linestyle="-.", linewidth=2, marker="o")

# Add temperature labels above each point 
for i in range(len(Temperature)):
    if Temperature[i] != max_temp:
        plt.text(Time[i], Temperature[i] + 1.5, str(Temperature[i]),
                 fontsize=10, ha="center")

# Annotate the highest temperature 
plt.annotate(f"Highest Temperature: {max_temp}",
             xy=(max_time, max_temp),  # Pointing to the actual max point
             xytext=(Time[-2], max_temp + 3),  # Text near 11AM, above the point
             arrowprops={"facecolor": "blue", "shrink": 0.05},
             fontsize=10, color="blue")

# Annotate the lowest temperature
plt.annotate(f"Lowest Temperature: {min_temp}",
             xy=(min_time, min_temp),  # Pointing to the lowest temp
             xytext=(min_time, min_temp - 4),  # Text below the point
             arrowprops={"facecolor": "purple", "shrink": 0.05},
             fontsize=10, color="purple")

# Add average temperature reference line
avg_temp = sum(Temperature) / len(Temperature)
plt.axhline(y=avg_temp, color='gray', linestyle='--',
            label=f'Avg Temp: {avg_temp:.1f}°C')

# Axis labels and title
plt.xlabel("Time", fontsize=12)
plt.ylabel("Temperature (°C)", fontsize=12)
plt.title("Temperature Changes in Morning", fontsize=14, fontweight="bold")

# Set y-axis limit for better spacing
plt.ylim(10, 30)

# Enable grid with custom style
plt.grid(True, linestyle="-.", alpha=0.8)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend in top-left to avoid overlap with annotations
plt.legend(loc='upper left', framealpha=0.9)

# Adjust layout to fit elements
plt.tight_layout()

# Show the final plot
plt.show()
```

```python
# Question-6: A school conducted a survey on the favorite sports of students. The number of students preferring each sport is 
# given below. Create a bar chart to represent the data and determine which sport is the most popular.
```

```python
import matplotlib.pyplot as plt

# Data
Sport = ["Football", "Basketball", "Cricket", "Tennis", "Volleyball"]  # Categories
Total_Students = [50, 40, 70, 30, 45]  # Corresponding values

colors = ['purple', 'orange', 'blue', 'green', 'red']  # Colors for each bar
plt.style.use('ggplot')  # Set plot style for better aesthetics

# Plot bar chart with custom colors and bar width
plt.bar(Sport, Total_Students, color=colors, width=0.5)  # Bars with colors

# Add value labels on top of each bar for clarity
for i, value in enumerate(Total_Students):
    plt.text(i, value + 1, str(value), ha='center', fontsize=10)  # Label above bars

# Set title and axis labels with font sizes and bold title
plt.title("Most Popular Sports", fontsize=14, fontweight='bold')  # Chart title
plt.xlabel("Sport", fontsize=12)  # X-axis label
plt.ylabel("Total Number of Students", fontsize=12)  # Y-axis label

# Enable grid only on y-axis to reduce clutter, dashed and slightly transparent
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.xticks(rotation=15)  # Rotate x-axis labels slightly for better fit and readability

plt.tight_layout()  # Adjust layout to fit everything without overlap

plt.show()  # Display the plot
```

```python
# Question-7: Plot a side-by-side bar chart to compare the monthly sales of Product A and Product B over the 4 months. Use different colors for each product,add a legend, 
# and label the x-axis with month names.
```

```python
import matplotlib.pyplot as plt  # plotting library
import numpy as np               # numerical operations

Month = ["January", "February", "March", "April"]  # months list
Product_A = [150, 280, 220, 170]                   # sales data A
Product_B = [200, 210, 190, 230]                   # sales data B

plt.style.use("Solarize_Light2")  # set plot style

x = np.arange(len(Month))         # x positions [0,1,2,3]
width = 0.25                     # bar width

plt.bar(x, Product_A, width=width, label="Product A", color="red")  # bars for A
plt.bar(x + width, Product_B, width=width, label="Product B", color="yellow")  # bars for B shifted

plt.xlabel("Month")               # x-axis label
plt.ylabel("Number of Products") # y-axis label
plt.title("Monthly Sales")       # plot title

plt.grid(True, linestyle="--", alpha=0.5, color="gray")  # grid lines
plt.legend(loc="upper left", framealpha=0.5)            # legend with transparency

plt.xticks(x + width / 2, Month)  # x-ticks at center of bars

plt.tight_layout()                # adjust layout
plt.show()                       # display plot
```
