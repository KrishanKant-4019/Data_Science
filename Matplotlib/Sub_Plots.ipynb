# Basic Subplot

```python
import matplotlib.pyplot as plt  

x = [1, 2, 3, 4, 5]  

y1 = [i * 2 for i in x] 
y2 = [i ** 2 for i in x]  

# Create the first subplot in a 1 row x 2 columns grid, position 1
plt.subplot(1, 2, 1)  
plt.plot(x, y1)  # Plot x vs y1 on the first subplot
plt.title("Double of x")  

# Create the second subplot in a 1 row x 2 columns grid, position 2
plt.subplot(1, 2, 2)
plt.plot(x, y2)  # Plot x vs y2 on the second subplot
plt.title("Square of x")  

plt.tight_layout()  
plt.show()  
```

# 2×2 Grid of Subplots

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]                 

y1 = [i * 2 for i in x]              
y2 = [i ** 2 for i in x]             
y3 = [i ** 0.5 for i in x]        
y4 = [10 - i for i in x]             

plt.figure(figsize=(8, 6))           # Create a new figure with specified size (width=8, height=6 inches)

# --- First subplot ---
plt.subplot(2, 2, 1)                 # Create subplot grid (2 rows × 2 columns), activate subplot #1 (top-left)
plt.plot(x, y1)                     
plt.title("x * 2")                 

# --- Second subplot ---
plt.subplot(2, 2, 2)                 # Activate subplot #2 (top-right)
plt.plot(x, y2)                      
plt.title("square of x")           

# --- Third subplot ---
plt.subplot(2, 2, 3)                 # Activate subplot #3 (bottom-left)
plt.plot(x, y3)                     
plt.title("square root of x")   

# --- Fourth subplot ---
plt.subplot(2, 2, 4)                 # Activate subplot #4 (bottom-right)
plt.plot(x, y4)                      
plt.title("10 - x")                 

plt.tight_layout()                  
plt.show()                          
```

```python
import matplotlib.pyplot as plt              

x = [1, 2, 3, 4, 5]                            

y1 = [i * 2 for i in x]                         
y2 = [i ** 2 for i in x]                        

fig, axs = plt.subplots(1, 2, figsize=(8, 6))   # Create 1 row, 2 columns of subplots (side by side), size 8x6 inches

axs[0].plot(x, y1)                              # Plot y1 on the first subplot (left)
axs[0].set_title("x * 2")                     

axs[1].plot(x, y2)                              # Plot y2 on the second subplot (right)
axs[1].set_title("x ** 2")                      

plt.suptitle("Simple Comparison Plots", fontsize=14)  # Add a global title above all subplots
plt.tight_layout()                             

fig.subplots_adjust(top=0.85)                   # Add spacing at the top for the main title
fig.savefig('my_plots.png')                     # Save the entire figure as an image file

plt.show()                                   
```

 # Looping Over Subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))  # Create 2x2 grid of subplots with figure size 8x6 inches
ys = [y1, y2, y3, y4]                         # List of y-values for each subplot
titles = ['x * 2', 'x squared', 'sqrt(x)', '10 - x']  # Titles for each subplot

for i in range(2):                            # Loop over subplot rows (0 and 1)
    for j in range(2):                        # Loop over subplot columns (0 and 1)
        idx = i * 2 + j                       # Calculate linear index to access ys and titles
        axs[i, j].plot(x, ys[idx])            # Plot data on the subplot at row i, column j
        axs[i, j].set_title(titles[idx])      # Set title of the current subplot

plt.tight_layout()                          
plt.show()                                   
```
