```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Plotting of Vectors

```python
sns.set()
```

```python
# Single Vector
plt.quiver(0, 0, 4, 5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
# Multiple Vectors
plt.quiver(0, 0, 4, 5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'blue')
plt.quiver(0, 0, -3, -6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'yellow')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

# Addition of 2 Vectors

```python
vector_1 = np.asarray([0, 0, 2, 3])
vector_2 = np.asarray([0, 0, 3, -2])

sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'blue')
plt.quiver(0, 0, 3, -2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'yellow')
plt.quiver(0, 0, 5, 1, scale_units = 'xy', angles = 'xy', scale = 1, color = 'red')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

# Subtraction of 2 Vectors

```python
vector_1 = np.asarray([0, 0, 2, 3])
vector_2 = np.asarray([0, 0, 3, -2])

difference = vector_1 - vector_2
print(difference)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'blue')
plt.quiver(0, 0, 3, -2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'yellow')
plt.quiver(0, 0, -1, 5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'red')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

# Multiplying a Vector by a Scalar

```python
vector_1 = np.asarray([0, 0, 2, 3])

vector_2 = 2 * vector_1
print(vector_2)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'blue')
plt.quiver(0, 0, 4, 6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'red')
plt.xlim(0, 8)
plt.ylim(0, 8)
plt.show()
```

```python
vector_1 = np.asarray([0, 0, 2, 3])

vector_2 = -0.5 * vector_1
print(vector_2)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'blue')
plt.quiver(0, 0, -1, -1.5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'red')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

# Dot Product of 2 Vectors

```python
a = np.array([2, 3])
b = np.array([4, 4])

a_dot_b = np.dot(a, b)

print(a_dot_b)
```

```python
c = np.array([20, 40, 45])
d = np.array([54, 67, 23])

c_dot_d = np.dot(c, d)

print(c_dot_d)
```

# Cross Product of 2 Vectors

```python
a = np.array([2, 3])
b = np.array([4, 4])

a_cross_b = np.cross(a, b)

print(a_cross_b)
```

```python
c = np.array([5, 10, 20])
d = np.array([15, 20, 25])

c_cross_d = np.cross(c, d)

print(c_cross_d)
```

# Projection of Vector

```python
a = np.array([2, 4, 5])
v = np.array([7, 8, 6])

a_dot_v = np.dot(a, v)

magnitude_of_v = np.sqrt(sum(v**2))

proj_of_a_on_v = (a_dot_v / magnitude_of_v ** 2) * v

print("Projection of a vector on v vector: ", proj_of_a_on_v)
```
