# Opening a CSV file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

response = requests.get(url)
data = StringIO(response.text)

pd.read_csv(data).head()
```

# Converters

```python
def rename(sex):
    if sex == "male":
        return "M"
    elif sex == "female":
        return "F"
    else:
        return sex
```

```python
pd.read_csv("titanic.csv", converters = {"Sex": rename})
```

# na_values parameter

```python
pd.read_csv("titanic.csv", na_values = ["male"])
```
