```python
# Importing Module
import requests 
```

# Adding Headers to Mimic a Browser

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"  # Imitates a real browser to avoid being blocked by the website
}
```

# Sending a Basic GET Request

```python
for i in range(1, 51):  # Loop through page numbers 1 to 50
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"  # Format the URL for each page
    a = requests.get(url, headers=headers)  # Send GET request with custom headers
    
    with open(f"HTML/page{i}.html", "w", encoding="utf-8") as f:  # Open a file to save the HTML content
        f.write(a.text)  # Write the page content to the file
        print(f"Downloaded page {i} successfully")  # Print success message
```

 # Checking the Response Status

```python
print(a.status_code)  # prints the HTTP status code from the response object 'a'
```

```python
if a.status_code == 200:
    print("Page fetched successfully!")
else:
    print("Failed to fetch the page!")
```

# Handling Connection Errors

```python
try:
    response = requests.get(url, timeout=5)         # Send GET request with 5s timeout
    response.raise_for_status()                     # Raise error for bad status codes (4xx/5xx)
    print(response.text)                            # Print the response content
except requests.exceptions.RequestException as e:   # Catch any request-related exception
    print(f"An error occurred: {e}")                # Print the error message
```
