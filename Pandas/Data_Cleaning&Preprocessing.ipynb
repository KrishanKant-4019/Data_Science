```python
import pandas as pd
```

```python
df = pd.read_csv("data_cleaning.csv")
```

```python
df
```

# Handling Missing Values

```python
# Check for Missing Data(NaN -> Not a Number)
df.isnull() 
```

```python
df.isnull().sum() # Counts null value in each columns
```

```python
# Drop rows with missing data
df.dropna() 
```

```python
df.dropna(axis=1) # Drop columns with missing value
```

```python
# Fill Missing Data
df.fillna(0)
```

```python
df["Age"].fillna(df["Age"].mean()) # Replace NaN with mean
```

```python
df
```

```python
df.ffill() # Forward fill
```

```python
df.bfill() # Backward fill
```

#Detecting & Removing Duplicates

```python
df.duplicated() # Check for duplicates
```

```python
df.drop_duplicates() # Drop duplicate rows
```

```python
df.duplicated(subset=["Age","Name"]) # Check duplicate values on specific columns
```

#String Operations with ".str"

```python
df["Name"].str.lower() # Converts all names to lowercase
```

```python
df["City"].str.contains("delhi", case = False) # Checks if 'delhi' is in the city name, case-insensitive
```

```python
df["Email"].str.split("@") # Return a pandas Series where each element is a list of strings (the split parts). This is where a Python list comes into play, but the outer object is still a pandas Series.
```

# Type Conversions with "astype()"

```python
df2=df.dropna().copy()
```

```python
df2["Age"] = df2["Age"].astype(int) # Converting float to integer
```

```python
df2
```

```python
df2.info() # Information about DataFrame
```

```python
df2.dtypes # Check data types
```

#Applying Functions

```python
# apply() -> Apply any function to rows or columns
df["Age Group"] = df["Age"].apply(lambda x: "Adult" if x >= 18 else "Minor")
```

```python
df
```

```python
# map() -> Element-wise mapping for Series
gender_map = {"M": "Male", "F": "Female"}
df["Gender"] = df["Gender"].map(gender_map)
```

```python
df
```

```python
# replace() -> Replace specific values
df["City"].replace({"Del": "Delhi", "Mum": "Mumbai"})
```
