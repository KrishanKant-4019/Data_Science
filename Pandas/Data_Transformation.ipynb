# Sorting & Ranking

```python
# Sort by Values(Ascending by default)
df.sort_values("Year") 
```

```python
df.sort_values("Year", ascending = False) # Descending order
```

```python
# Sorting of multiple columns
df.sort_values(["Year", "IMDb"]) 
```

```python
# Reset Index
df.reset_index()
```

```python
df.reset_index(drop = True, inplace = True) # Reset the index and drop the old index
```

```python
df
```

```python
df.sort_index()
```

```python
# Ranking
df["Rank"] = df["IMDb"].rank(ascending = False, method = "dense")
```

```python
df
```

# Renaming Columns & Index

```python
df.rename(columns = {"IMDb" : "Ratings"}, inplace = True) # Renaming of columns
```

```python
df
```

```python
df.rename(index = {0 : "row1", 1 : "row2"}, inplace = True) # Renaming of indices
```

```python
df
```

```python
df.columns = ["Hero", "Movie", "Annum", "Category", "Total Collection", "Reviews", "Position"] # Renaming of all columns 
```

# Changing Column Order

```python
df = df[["Hero","Annum","Movie","Category","Reviews","Total Collection"]] # Reordering of columns
```

```python
df
```

```python
cols = ["Category"] + [col for col in df.columns if col != "Category"] # Moving one column to the front
df = df[cols]
```
