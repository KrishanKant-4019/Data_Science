```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
# Plot Passenger Age Distribution
# 📁 Dataset: titanic
# 📊 Function: histplot(), kdeplot()

# Q: Create a histogram of the age column with 30 bins and overlay a KDE curve.

plt.figure(figsize = (10, 6))

sns.histplot(
    data = titanic,
    x = "age",
    bins = 30,
    kde = True,
    stat = 'density', 
    color = 'skyblue', 
    edgecolor = 'black',
    alpha = 0.7
)
plt.title("Passenger Age Distribution on Titanic", fontsize = 16)
plt.xlabel("Age")
plt.ylabel("Density")
plt.grid(True, alpha = 0.7, color = "gray")
plt.tight_layout()
plt.show()
```

```python
# 2. Visualize Gender Counts
# 📁 Dataset: titanic
# 📊 Function: countplot()

# Q: Create a bar chart showing the number of passengers by gender (sex column).

sns.set_style("whitegrid")
sns.set_palette("pastel")

plt.figure(figsize = (8, 6))

ax = sns.countplot(data = titanic,
              x = "sex",
              edgecolor = "black",
              alpha = 0.85
             )
plt.title("Passenger Gender Count on Titanic", fontsize = 16, fontweight = "bold")
plt.xlabel("Sex")
plt.ylabel("Number of Passengers")

for p in ax.patches:
    count = p.get_height()
    ax.annotate(f"{count}", xy = (p.get_x() + p.get_width() / 2, count),
                ha = "center", va = "bottom", fontsize = 12, fontweight = "bold")

plt.grid(True, alpha = 0.7)
plt.tight_layout()
plt.show()
```
