```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

# Plot Styles and Themes

```python
import matplotlib.pyplot as plt

for style in ["darkgrid", "whitegrid", "dark", "white", "ticks"]:
    sns.set_style(style)
    sns.histplot(titanic["age"].dropna(), kde = True)
    plt.title(f"Style : {style}")
    plt.show()
```

# Color Palettes

```python
palettes = ['deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind']

for palette in palettes:
    sns.set_palette(palette)
    sns.countplot(x = "class", data = titanic)
    plt.title(f"Palette : {palette}")
    plt.show()
```

 # Distribution Plots

```python
# histplot()
sns.histplot(titanic["age"].dropna(), bins = 30, kde = True, color = "purple")
plt.title("Age Distribution")
plt.show()
```

```python
# kdeplot()
sns.kdeplot(titanic["fare"], fill = True)
plt.title("Kernel Density Estimate of Fare")
plt.show()
```

 # Categorical Plots

```python
# countplot()
sns.countplot(x = "sex", hue = "sex", data = titanic, palette = "Set2", legend = False )
plt.title('Count of Passengers by Sex')
plt.show()
```

```python
# boxplot()
sns.boxplot(x = "class", hue = "class", y = "age", data = titanic, palette = "pastel", legend = False)
plt.title("Age Distribution by Passenger Class")
plt.show()
```

```python
# violinplot()
sns.violinplot(x = "survived", hue = "survived", y = "fare", data = titanic, palette = "muted")
plt.title("Fare Distribution by Survival Status")
plt.show()
```

```python
# stripplot()
sns.stripplot(x = "class", y = "age", data = titanic, jitter = True)
plt.title("Age by Class - Stripplot")
plt.show()
```

```python
# swarmplot()
sns.swarmplot(x = "class", y = "age", data = titanic, size = 3)
plt.title("Age by Class - Swarmplot")
plt.show()
```
