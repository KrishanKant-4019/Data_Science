# Setup and Load Data

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_theme(style = "darkgrid")

df = pd.read_csv("sales_data.csv")
print(df.head())
```

# Line plot: Laptop Revenue Trend

```python
laptop_data = df.query("Product == 'Laptop'")
sns.lineplot(data = laptop_data, x = 'Month', y = 'Revenue', marker = 'o', color = 'blue')
plt.title("Laptop Revenue Trend", fontsize = 16)
plt.tight_layout()
plt.show()
```

# Bar plot: April Revenue by Product

```python
april_data = df.query("Month == 'Apr'")

sns.barplot(data = april_data, x = 'Product', y = 'Revenue', 
            hue = 'Product', palette = 'pastel', legend = False)
plt.title("April Revenue by Product", fontsize = 16)
plt.tight_layout()
plt.show()
```

```python
# Boxplot: Revenue distribution by Product
sns.boxplot(data = df, x = 'Product', y = 'Revenue', 
            hue = 'Product', palette = 'Set2', legend = False)
plt.title("Revenue distribution by Product", fontsize = 16)
plt.tight_layout()
plt.show()
```

# Violin plot: Profit distribution by Product

```python
sns.violinplot(data=df, x='Product', y='Profit', 
               hue = 'Product', palette='muted', legend = False)
plt.title("Profit Distribution by Product", fontsize = 16)
plt.tight_layout()
plt.show()
```

# Scatter plot: Units Sold vs Profit

```python
sns.scatterplot(data=df, x='Units_Sold', y='Profit', hue='Product', s=100)
plt.title("Units Sold vs Profit by Product", fontsize = 16)
plt.tight_layout()
plt.show()
```

# Heatmap: Correlation Matrix

```python
corr = df[['Units_Sold', 'Revenue', 'Profit']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix", fontsize = 16)
plt.tight_layout()
plt.show()
```

# Pairplot: Visualizing Pairwise Relationships

```python
sns.pairplot(df, hue='Product', height=2.5)
plt.suptitle("Pairwise Relationships in Sales Data", y=1.02)
plt.tight_layout()
plt.show()
```

# Customizing Aesthetics

```python
sns.set_theme(style = "whitegrid", palette = "bright")

sns.barplot(data = april_data, x = 'Product', y = 'Revenue')
plt.title("April Revenue by Product - Bright Palette", fontsize = 16)
plt.tight_layout()
plt.show()
```
