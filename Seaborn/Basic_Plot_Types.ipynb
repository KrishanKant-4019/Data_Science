```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
tips = sns.load_dataset("tips")            # Load the 'tips' dataset
sns.lineplot(x="total_bill", y="tip", data=tips)  # Plot line graph of total_bill vs tip
plt.title("Line Plot")                      # Set the plot title
plt.show()                                  # Display the plot
```

# Scatter Plot

```python
tips = sns.load_dataset("tips")            # Load the tips dataset
sns.lineplot(x="total_bill", y="tip", data=tips)  # Plot line plot: total_bill vs tip
plt.title("Line Plot")                      # Add title to the plot
plt.show()                                  # Show the plot
```

# Bar Plot

```python
sns.barplot(x="day", y="total_bill", data=tips)  # Create bar plot of average total_bill per day
plt.title("Average Bill per Day")                # Set the plot title
plt.show()                                       # Display the plot
```

# Box Plot

```python
sns.boxplot(x="day", y="total_bill", data=tips)  # Draw boxplot of total_bill distribution for each day
plt.title('Boxplot of Total Bill per Day')       # Add title to the plot
plt.show()                                       # Display the plot
```

# Heatmap (Correlation Matrix)

```python
import matplotlib.pyplot as plt               # Import matplotlib for plotting

flights = sns.load_dataset('flights')        # Load the 'flights' dataset
pivot_table = flights.pivot(index="month", columns="year", values="passengers")  
# Reshape data: months as rows, years as columns, values are passengers

sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")  
# Plot heatmap with passenger counts annotated as integers, using Yellow-Green-Blue color map

plt.title('Heatmap of Passengers')            # Add title to the heatmap
plt.show()                                    # Show the heatmap plot
```
